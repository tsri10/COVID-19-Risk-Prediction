```python
import pandas as pd

df1=pd.read_csv('diabetes.csv')
df2=pd.read_csv('Covid Dataset.csv')
df3=pd.concat([df1, df2], axis=1)
df=pd.DataFrame(df3)
df.dropna(inplace=True)
df
```

```python
df.drop(df.columns[[0,3,16,17,18,19,20,21,24,25,26,27,28]], axis = 1, inplace = True)
df
```

```python
# @title Blood Pressure vs BMI by Diabetes Status

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['BMI'], df['BloodPressure'], c=df['Outcome'], cmap='viridis')
plt.xlabel('BMI')
plt.ylabel('Blood Pressure')
plt.title('Blood Pressure vs BMI by Diabetes Status')
_ = plt.colorbar(label='Diabetes Outcome')
```

```python
# @title Blood Pressure vs BMI by Glucose level

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['BMI'], df['BloodPressure'], c=df['Glucose'], cmap='viridis')
plt.xlabel('BMI')
plt.ylabel('Blood Pressure')
plt.title('Blood Pressure vs BMI by Glucose level')
_ = plt.colorbar(label='Glucose')
```

```python
result=[]
def result(row):
  if(row['Outcome']==1.0 and (row['Glucose']>=140 or row['Age']>=50.0)):
    if(row['Breathing Problem']+row['Fever']+row['Dry Cough']+row['Sore throat']+row['Running 0se']+row['Asthma']+row['Chronic Lung Disease']+row['Abroad travel']+row['Contact with COVID Patient']>=3):
      return "More Likely"
  else:
    return "Less Likely"

df['Result'] = df.apply(result, axis=1)
df.to_csv('Result.csv')
```

```python
from sklearn.model_selection import train_test_split


y = df.pop('Result')
X = df

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
